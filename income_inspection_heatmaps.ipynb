```python
main_df = pd.read_csv('all_data_cleaned.csv')
main_df.head()
```

```python
# group by zip, to make heat maps
zip_grp = main_df.groupby('zipcode', as_index=False)
zip_df = zip_grp.mean()
zip_df
sf_zip_df = zip_df[(zip_df.zipcode != 90012) &
                   (zip_df.zipcode != 94301) &
                   (zip_df.zipcode != 94619) &
                   (zip_df.zipcode != 94931) &
                   (zip_df.zipcode != 94019)]
sf_zip_df.head()
```

```python
locations = sf_zip_df[['zip_lat','zip_lng']]
inspection_scores = sf_zip_df['average_health_score']
income = sf_zip_df['median_income']
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=inspection_scores, 
                                 dissipating=True,
                                 point_radius=30)

# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```
